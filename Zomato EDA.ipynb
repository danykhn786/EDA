```python
df_cc = pd.read_excel("Country-Code.xlsx")
df_cc.head()
```

```python
final_df = pd.merge(df,df_cc,on = 'Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
final_df.shape
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#Creating Pie Chart - Top 3 Countries using zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: 
Zomato's Maximum records or transactions are from India followed by USA and United Kingdom.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
<ol>
<li> When rating are in between 4.5 to 4.9 the rating is excellent.</li>

<li> When ratings are in betweenn 4.0 to 4.4 the rating is very good.</li>

<li> When ratings are between 3.5 to 3.9 the rating is good.</li>

<li> When ratings are in between 2.5 to 3.4 the rating is average.</li>

<li> When ratings are in between 1.8 to 2.4 the rating is poor.</li>
</ol>

```python
ratings.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',palette=['blue','red','orange','yellow','green','green'],data = ratings)
```

Observation: 
<ol>
    <li>Not rated count is very high.</li>
    <li>Maximum ratings are in between 2.5 to 3.4</li>
</ol>

```python
sns.countplot(x='Rating color',data=ratings,palette = ['blue','red','orange','yellow','green','green'])
```

```python
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

Observations:
